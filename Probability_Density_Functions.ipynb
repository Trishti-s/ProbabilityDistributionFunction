```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
!unzip /data.csv.zip
```

```python
df = pd.read_csv("data.csv", encoding='latin1')
```

```python
print(df.columns)
```

```python
x = df['no2'].dropna().values
```

```python
r=102313056

ar=0.05*(r%7)
br=0.3*((r%5)+1)

print("ar:",ar)
print("br:",br)
```

```python
z=x+ar*np.sin(br*x)
```

```python
plt.hist(z,bins=50,density=True)
plt.title("Histogram of Tranformed variable Z")
plt.show()
```

```python
mu=np.mean(z)
```

```python
variance=np.var(z)
```

```python
lambda_val=1/(2*variance)
```

```python
c=1/np.sqrt(2*np.pi*variance)
```

```python
def pdf(z,mu,lam,c):
  return c*np.exp(-lam*(z-mu)**2)
z_range=np.linspace(min(z),max(z),500)
p_pred=pdf(z_range,mu,lambda_val,c)
```

```python
plt.hist(z,bins=50,density=True,alpha=0.5)
plt.plot(z_range,p_pred)
plt.title("estimated PDF")
plt.show()
```

```python
print("mu:",mu)
print("lambda:",lambda_val)
print("c:",c)
```
